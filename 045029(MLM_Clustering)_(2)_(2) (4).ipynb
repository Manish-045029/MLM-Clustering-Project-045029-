## **Data dimensions**

The dataset consists of 70,000 entries and 13 columns.
Each row represents an individual's health-related information.
The columns include:

1. Age: The age of the individual in days (integer).
2. Height: The height of the individual in centimeters (integer).
3. Weight: The weight of the individual in kilograms (float).
4. Gender: The gender of the individual, represented as a categorical code.
5. Systolic blood pressure (ap_hi): The systolic blood pressure measured during examination (integer).
6. Diastolic blood pressure (ap_lo): The diastolic blood pressure measured during examination (integer).
7. Cholesterol: The cholesterol level of the individual, categorized as 1 for normal, 2 for above normal, and 3 for well above normal.
8. Glucose: The glucose level of the individual, categorized as 1 for normal, 2 for above normal, and 3 for well above normal.
9. Smoking: Whether the individual smokes or not, represented as binary (0 for non-smoker, 1 for smoker).
10. Alcohol intake: Whether the individual consumes alcohol or not, represented as binary (0 for non-drinker, 1 for drinker).
11. Physical activity: Whether the individual engages in physical activity or not, represented as binary (0 for inactive, 1 for active).
12. Presence or absence of cardiovascular disease (cardio): The target variable indicating the presence or absence of cardiovascular disease, represented as binary (0 for absence, 1 for presence).

## **Analysis by K-means:**

**a.) Elbow Curve Analysis:**

An elbow curve was plotted to determine the optimal number of clusters for K-means clustering.

The elbow curve suggests that the optimal number of clusters might be 3, as the curve starts to flatten beyond this point.

**b.) K-means Clustering:**

K-means clustering was performed with different numbers of clusters (2, 3, 4, and 5) to evaluate the clustering performance.

**c.) Performance Metrics:**


**SS (Silhouette Score):**

The Silhouette Score measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation).

Higher Silhouette Scores indicate better clustering performance.

**DB (Davies-Bouldin) Score**:

The Davies-Bouldin Score measures the average similarity between each cluster
and its most similar cluster.

Lower Davies-Bouldin Scores indicate better clustering performance.
Performance Results:

## **K-mean Clustering**

**Elbow Curve**

```python
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(df_1)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

## **Chi-Square**

```python
from scipy.stats import chi2_contingency

# Extracting categorical variables
cat_variables = ['gender_y',	'cholesterol_y',	'gluc_y']

# Perform Chi-square test for each categorical variable
chi2_results_cat = {}
for column in cat_variables:
    contingency_table = pd.crosstab(df_1_kmcluster[column], km_3cluster_model)
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    chi2_results_cat[column] = {'Chi-square': chi2, 'P-value': p_value}

# Print Chi-square test results for categorical variables
for column, result in chi2_results_cat.items():
    print(f"Variable: {column}")
    print(f"Chi-square: {result['Chi-square']}")
    print(f"P-value: {result['P-value']}")
    print()
```
